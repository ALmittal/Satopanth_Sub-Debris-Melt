```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'F:\Biases.csv')
df.head()
```

```python
day = df['Day']
bias_1D = df['1D']
bias_2D = df['2D']
bias_5D = df['5D']
bias_10D = df['10D']
bias_15D = df['15D']
plt.rcParams['figure.dpi'] = 180

plt.plot(day, bias_1D, label='1 Day', color='#FF80ED', linewidth=1)
plt.plot(day, bias_2D, label='2 Day', color='#8FF7FE', linewidth=1)
plt.plot(day, bias_5D, label='5 Day', color='#63FE83', linewidth=1)
plt.plot(day, bias_10D, label='10 Day', color='#5964FF', linewidth=1)
plt.plot(day, bias_15D, label='15 Day', color='r', linewidth=2)
plt.legend()
plt.xlabel('Julian day of 2017')
plt.ylabel('Temperature Bias in Â°C')
```
