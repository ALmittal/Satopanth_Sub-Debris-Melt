```python
import pandas as pd
import matplotlib.pyplot as plt 
df = pd.read_csv(r'F:\Abhilah_First\Results\Era5Land\Total_periodic_melt.csv')
df
```

```python
julian_day_intervals = []
for i, j in zip(df['Jd1'], df['Jd2']):
    julian_day_intervals.append(str(i) + '-' + str(j))

julian_day_intervals
```

```python
plt.rcParams['figure.dpi'] = 110
plt.plot(julian_day_intervals, df['Observed Melt'], label ='Observed')
plt.plot(julian_day_intervals, df['Model 1'],label ='Model 1')
plt.plot(julian_day_intervals, df['Model 2'],label ='Model 2')
plt.plot(julian_day_intervals, df['Model 3'],label ='Model 3')
plt.legend()
plt.xlabel('julian day of 2017')
plt.ylabel('Total melt (m^3 10^6)')
```
