```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import random
df_sp = pd.read_csv(r'F:\Abhilah_First\Results\Era5Land\noised_stake_data.csv')
df_sp.head()
```

```python
df_model3_param = pd.read_csv(r'F:\Abhilah_First\Results\Era5Land\Model3param.csv')
df_model3_param
```

```python
ddf_noised = [] # 1000 ddf
d0_noised = [] #1000 d0
a_noised = [] # 1000 a
for i in range(1000):
    ddf_noised.append(df_model3_param['DDF'][0] + random.uniform(-df_model3_param['Uncertainty of DDF'][0],df_model3_param['Uncertainty of DDF'][0]))
    d0_noised.append(df_model3_param['d0'][0] + random.uniform(-df_model3_param['Uncertainty of d0'][0],df_model3_param['Uncertainty of d0'][0]))
    a_noised.append(df_model3_param['a'][0] + random.uniform(-df_model3_param['Uncertainty of a'][0],df_model3_param['Uncertainty of a'][0]))
df_all_iterations = pd.DataFrame(ddf_noised, columns = ['ddf'])
df_all_iterations['d0'] = d0_noised
df_all_iterations['a'] = a_noised
MSEs = [] # it has all the Mean sq errors for each  TF1 and TF2
weight = []
observed_melt_rate = [] #cm/day
pdd = list(df_sp['pdd'])
db_thickness = list(df_sp['Derbis thickness (cm)'])
ob_day = list(df_sp['obs_period(days)'])

for ob, m in zip(df_sp['obs_period(days)'], df_sp['melt (cm)']):
    observed_melt_rate.append(m/ob)
    
for ddf, d0, a in zip(ddf_noised, d0_noised, a_noised):
    model_melt_rate_sqd = []
    for ob,mr,p,db in zip(ob_day, observed_melt_rate, pdd, db_thickness):
        model_melt_rate_sqd.append((mr - ddf*p/((1+(db/d0)**2)**a)/10/ob)**2)
    MSEs.append(sum(model_melt_rate_sqd)/len(model_melt_rate_sqd))
    weight.append(math.exp(-sum(model_melt_rate_sqd)/len(model_melt_rate_sqd)))
df_all_iterations['MSE'] = MSEs
df_all_iterations['Weight'] = weight
df_all_iterations
```

```python
print(min(weight), max(weight))
count = 0
for i in weight:
    if i <= 0.655:
        count +=1
print('count: ', count) 
# there is no zero weight, number of non zero weights is 1000
```

```python
df_all_iterations['ddf*Weight'] = df_all_iterations['ddf']*df_all_iterations['Weight']
df_all_iterations['d0*Weight'] = df_all_iterations['d0']*df_all_iterations['Weight']
df_all_iterations['a*Weight'] = df_all_iterations['a']*df_all_iterations['Weight']

df_all_iterations['Weight*(ddf - xBARw)^2'] = df_all_iterations['Weight']*(df_all_iterations['ddf'] - sum(df_all_iterations['ddf*Weight']/1000) )**2
df_all_iterations['Weight*(d0 - yBARw)^2'] = df_all_iterations['Weight']*(df_all_iterations['d0'] - sum(df_all_iterations['d0*Weight']/1000) )**2
df_all_iterations['Weight*(a - zBARw)^2'] = df_all_iterations['Weight']*(df_all_iterations['a'] - sum(df_all_iterations['a*Weight']/1000) )**2

df_all_iterations
```

```python
ddf_variability = (sum(df_all_iterations['Weight*(ddf - xBARw)^2'])*(1000)/(1000*(1000-1)*sum(df_all_iterations['Weight'])) )**(0.5)


d0_variability = (sum(df_all_iterations['Weight*(d0 - yBARw)^2'])*(1000)/(1000*(1000-1)*sum(df_all_iterations['Weight'])) )**(0.5)

a_variability = (sum(df_all_iterations['Weight*(d0 - yBARw)^2'])*(1000)/(1000*(1000-1)*sum(df_all_iterations['Weight'])) )**(0.5)

print(df_all_iterations['ddf'][MSEs.index(min(MSEs))], df_all_iterations['d0'][MSEs.index(min(MSEs))], df_all_iterations['a'][MSEs.index(min(MSEs))])
print(round(ddf_variability,3), round(d0_variability,3), round(a_variability,3))
```
