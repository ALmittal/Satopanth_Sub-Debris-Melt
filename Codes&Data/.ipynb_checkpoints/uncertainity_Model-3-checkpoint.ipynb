```python
#model2
#using observation data
#DDF = 5.768
#d0 = 9.652
```

```python
ddf_range = [] #it has a series of ddfs from 4 to 7 (0.05 interval)
d0_range = [] #it has a series of TF2s from 8 to 11 (0.05 interval)
for i in np.arange(4, 7, 0.05):
    ddf_range.append(round(i, 2))

for i in np.arange(8, 11, 0.05):
    d0_range.append(round(i, 2))
```

```python
MSEs = [] # it has all the Mean sq errors for each  ddf and tf2
observed_melt_rate = [] #cm/day
pdd = list(df_sp['PDD'])
db_thickness = list(df_sp['debris_thickness(cm)'])
ob_day = list(df_sp['obs_period(days)'])

for ob, m in zip(df_sp['obs_period(days)'], df_sp['melt(cm)']):
    observed_melt_rate.append(m/ob)
weight = [] #exp(-mse)
data = []  
for ddf in ddf_range:
    for d0 in d0_range:
        data1 = []
        model_melt_rate_sqd = []
        for ob,mr,p,db in zip(ob_day, observed_melt_rate, pdd, db_thickness):
            model_melt_rate_sqd.append((mr - ddf*p/(1 + db/d0)/10/ob)**2)
        MSEs.append(sum(model_melt_rate_sqd)/len(model_melt_rate_sqd))
        weight.append(math.exp(-sum(model_melt_rate_sqd)/len(model_melt_rate_sqd)))
        data1.append(ddf)
        data1.append(d0)
        data1.append(sum(model_melt_rate_sqd)/len(model_melt_rate_sqd))
        data.append(data1)
      
df_all_iterations = pd.DataFrame(data, columns = ['DDF', 'd0', 'MSE'])
df_all_iterations['Weight'] = weight
df_all_iterations
```

```python
count = 0
for i in weight:
    if i <= 0.1:
        count +=1
max(weight)
```

```python
df_all_iterations['DDF*Weight'] = df_all_iterations['DDF']*df_all_iterations['Weight']
df_all_iterations['d0*Weight'] = df_all_iterations['d0']*df_all_iterations['Weight']

df_all_iterations['Weight*(DDF - xBARw)^2'] = df_all_iterations['Weight']*(df_all_iterations['DDF'] - sum(df_all_iterations['DDF*Weight']/3600) )**2
df_all_iterations['Weight*(d0 - yBARw)^2'] = df_all_iterations['Weight']*(df_all_iterations['d0'] - sum(df_all_iterations['d0*Weight']/3600) )**2

df_all_iterations
```

```python
ddf_variability = (sum(df_all_iterations['Weight*(DDF - xBARw)^2'])*3600/(3600*(3600-1)*sum(df_all_iterations['Weight'])) )**(0.5)


d0_variability = (sum(df_all_iterations['Weight*(d0 - yBARw)^2'])*3600/(3600*(3600-1)*sum(df_all_iterations['Weight'])) )**(0.5)


print(round(ddf_variability,2), round(d0_variability,2))
```
