```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
df_day_9tiles = pd.read_csv('F:\DG\Era5Land\Dunagiri_Hourly_T_1986.csv')
df_day_9tiles
```

```python
elevations = [4284.205724, 4298.853308, 
              4626.426531, 4041.854803, 
              4558.310148, 4995.68903, 
              4323.948817, 4979.197695, 5175.864546]
```

```python
lapse_rate_hourly = []
for i in range(4656):
    def func(X, a, c):
        return a*X + c
    
    # X and Ydata
    X = elevations
    Ydata = list(df_day_9tiles.iloc[i, 1:])
    popt, pcov = curve_fit(func, X, Ydata)
    #print(popt)
    lapse_rate_hourly.append(popt[0]*(-1000))
    
```

```python
sum(lapse_rate_hourly)/len(lapse_rate_hourly)
```

```python
df_day_9tiles['LapseRate'] = lapse_rate_hourly
df_day_9tiles
```

```python
continuous_day = []

for i in range(4656):
    cd = (df_day_9tiles.iloc[i, 0] - int(df_day_9tiles.iloc[i, 0]))*100/24 + int(df_day_9tiles.iloc[i, 0])
    continuous_day.append(cd)
```

```python
day_for_ave_lapserate = 2
window_size = int(day_for_ave_lapserate*24)
numbers_series = pd.Series(lapse_rate_hourly)
windows = numbers_series.rolling(window_size)
moving_averages = windows.mean()
moving_averages_list = moving_averages.tolist()
without_nans_lr = moving_averages_list[window_size - 1:]
df_day_9tiles['Moving_average_lapse_rate'] = moving_averages_list

plt.plot(df_day_9tiles['continuous_day'], lapse_rate_hourly)
plt.plot(df_day_9tiles['continuous_day'] ,df_day_9tiles['Moving_average_lapse_rate'])
```

```python
df = df_day_9tiles.iloc[48:, :] #df starts with day 154
df
```

# Calculating T at 3800 meter height

```python
df['T_3800']= df['Moving_average_lapse_rate']*(elevations[0] - 3800 )/1000 + df['30.7,79.8']
df.index = np.arange(4608)
df
```

```python
x = list(df['Julian day']).index(198.01)
y = list(df['Julian day']).index(214.01)
t_mean = sum(df['T_3800'][x:y])/(y-x)
t_mean
```

```python
dg_data_1986 = pd.read_csv('F:\DG\Era5Land\dg_glacier_data_1986.csv')
dg_data_1986
```

```python
t_mean_Era = []
for i in range(5):
    x = list(df['Julian day']).index(dg_data_1986['jd1'][i] + .01)
    y = list(df['Julian day']).index(dg_data_1986['jd2'][i] + 1.01)
    t_mean_Era.append(sum(df['T_3800'][x:y])/(y-x))
dg_data_1986['T_mean_Era'] = t_mean_Era
dg_data_1986['Bias'] = dg_data_1986['T_mean'] - dg_data_1986['T_mean_Era']
dg_data_1986
```
