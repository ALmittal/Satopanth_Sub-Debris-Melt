```python
f_histo = open('F:\DG\Era5Land\dg_catchment_histo.txt')
elevation = []
area = []
for line in f_histo:
    elevation.append(int(line.split()[0]))
    area.append(int(line.split()[1])*900/1000000)
df_histo = pd.DataFrame({'z':elevation,
                        'area':area})
df_histo
```

```python
plt.bar(elevation, area, width = 10)
```

# Finding Pdd

```python
area_weighted_pdd = []
for i in range(5):  #number of time intervals is 5
    areaXpdd = []   # for each time interval, area*pdd for different elevations of the histogram
    for a, z in zip(area, elevation):  
        y = list(df_T_lr_1986['Julian day']).index(df_bias['jd2'][i] + 1.01) #indices of starting and ending day
        x = list(df_T_lr_1986['Julian day']).index(df_bias['jd1'][i] + .01)
        temp_at_z = list(df_T_lr_1986['Moving_average_lapse_rate'][x:y]*(4284.205724 - z )/1000 + df_T_lr_1986['30.7,79.8'][x:y] + df_bias['Bias'][i])
        PDD = sum([s for s in temp_at_z if s > 0]) / 24 
        areaXpdd.append(PDD*a)
    area_weighted_pdd.append(sum(areaXpdd)/sum(area))
area_weighted_pdd
```

```python
df_bias['pdd_area_weighted'] = area_weighted_pdd
df_bias
```
