```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
df_sp = pd.read_csv(r'C:\Users\Abhilash Mittal\Desktop\All_T_data\SP_glacier_data.csv')
df_sp
```

```python
#model1
#using observation data
#ddf = DDF = 8.694; 
#TF2 = -0.526
```

```python
ddf_range = [] #it has a series of ddfs from 7 to 10 (0.05 interval)
tf2_range = [] #it has a series of TF2s from -2 to 1 (0.05 interval)
for i in np.arange(7, 10, 0.05):
    ddf_range.append(round(i, 2))

for i in np.arange(-3, 0, 0.05):
    tf2_range.append(round(i, 2))
```

```python
MSEs = [] # it has all the Mean sq errors for each  ddf and tf2
observed_melt_rate = [] #cm/day
pdd = list(df_sp['PDD'])
db_thickness = list(df_sp['debris_thickness(cm)'])
ob_day = list(df_sp['obs_period(days)'])

for ob, m in zip(df_sp['obs_period(days)'], df_sp['melt(cm)']):
    observed_melt_rate.append(m/ob)
weight = [] #exp(-mse)
data = []  
for ddf in ddf_range:
    for tf2 in tf2_range:
        data1 = []
        model_melt_rate_sqd = []
        for ob,mr,p,db in zip(ob_day, observed_melt_rate, pdd, db_thickness):
            model_melt_rate_sqd.append((mr - ddf*p*db**tf2/10/ob)**2)
        MSEs.append(sum(model_melt_rate_sqd)/len(model_melt_rate_sqd))
        weight.append(math.exp(-sum(model_melt_rate_sqd)/len(model_melt_rate_sqd)))
        data1.append(ddf)
        data1.append(tf2)
        data1.append(sum(model_melt_rate_sqd)/len(model_melt_rate_sqd))
        data.append(data1)
      
df_all_iterations = pd.DataFrame(data, columns = ['DDF', 'TF2', 'MSE'])
df_all_iterations['Weight'] = weight
df_all_iterations
```

```python
count = 0
for i in weight:
    if i <= 0.01:
        count +=1
count 
```

```python
df_all_iterations.to_csv(r'C:\Users\Abhilash Mittal\Desktop\All_T_data\Model1_observed_T_Uncertainity.csv')
```

```python
df_all_iterations['DDF*Weight'] = df_all_iterations['DDF']*df_all_iterations['Weight']
df_all_iterations['TF2*Weight'] = df_all_iterations['TF2']*df_all_iterations['Weight']

df_all_iterations['Weight*(DDF - xBARw)^2'] = df_all_iterations['Weight']*(df_all_iterations['DDF'] - sum(df_all_iterations['DDF*Weight']/3600) )**2
df_all_iterations['Weight*(TF2 - yBARw)^2'] = df_all_iterations['Weight']*(df_all_iterations['TF2'] - sum(df_all_iterations['TF2*Weight']/3600) )**2

df_all_iterations
```

```python
ddf_variability = (sum(df_all_iterations['Weight*(DDF - xBARw)^2'])*3405/(3600*(3405-1)*sum(df_all_iterations['Weight'])) )**(0.5)


TF2_variability = (sum(df_all_iterations['Weight*(TF2 - yBARw)^2'])*3405/(3600*(3405-1)*sum(df_all_iterations['Weight'])) )**(0.5)


print(round(ddf_variability,2), round(TF2_variability,2))
```
